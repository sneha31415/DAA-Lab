## **Selection Sort**
### Algorithm :
- step1. Iterate over all the elements using the first loop `i`
- step2. In this first loop ` i`, set `minindex = i` everytime
- step3. Now a 2nd loop will iterate over from `j = i` to end of the array and find the minimum element from the array in `j to n - 1`
- step4. after the end of the 2nd loop, everytime `swap` `array[i]` and the `minimum most index elem`. <br>
    This way after every iteration the minimum most element from the rest of the array will come at the starting index i.e `i`
- repeat steps 2 and 3 `untill i == n - 2`

image.png

**Time Complexity**<br>
To find the minimum element from the array of ` N ` elements, ` N -1` comparisons are required. After putting the minimum element in its proper position, the size of an unsorted array reduces to `N - 1` and then ` N - 2` comparisons are required to find the minimum in the unsorted array.

Therefore
`N -1 + N -2 + N -3 + .... = N * (N - 1) / 2`
 comparisons and
 swaps result in the overall complexity of `O(N^2)`


**Code for selection sort**

```python
def selection_sort(array: list[int]):
    n = len(array)
    # after n-1 iterations, the last elem left is already sorted
    for i in range(n - 1):
        # 0 to (i-1) is the sorted part. So find min from i to n
        min_index = i
        for j in range(i, n):
            if array[j] < array[min_index]:
                min_index = j
        # now we have found the index of the minimum elem from the unsorted part. So swap
        array[i], array[min_index] = array[min_index], array[i]
```

```python
test_array = [3,1,6,3,8,9,10,-1,0,-5]
print("original array is", test_array)
selection_sort(test_array)
print("sorted array is", test_array)
```

## **Heap sort**

### Algorithm:
- Build the max heap from the unsorted array
- swap the first elem and the last elem
- apply heapify on `i = 0` to `n - 1 - i` th elem
- repeat step 2 and 3

### Time complexity
1) The heap building (step1) costs `O(n)` time complexity(rather than O(n log n)). This involes heapifying the non leaf nodes.

2)  Then we swap which takes constant time .
    Then we call the heapify func on each node which costs logn * n .Heapify operation has a time complexity of O(log n) in the worst case.

So time complexity = `n + n logn` => `nlogn`



**Code for Heapsort**

```python
# helper function which downheapifies
# this will heapify the i th position in the array
def heapify(arr: list[int], i: int, n: int):
    parent_index = i
    Lchild_index = 2 * i + 1
    Rchild_index = 2 * i + 2

    while Lchild_index < n:
        min_index = parent_index
        if (arr[Lchild_index] > arr[min_index]):
            min_index = Lchild_index
        if (Rchild_index < n and arr[Rchild_index] > arr[min_index]):
            min_index = Rchild_index
            # return if parent index is the min index
        if (min_index == parent_index):
            return
            # else swap the parent index elem with the min index elem
        arr[parent_index] , arr[min_index] = arr[min_index], arr[parent_index]

        # update the parent_index, Lchild_index and Rchild_index
        parent_index = min_index
        Lchild_index = 2 * parent_index + 1
        Rchild_index = 2 * parent_index + 2


# main function
def heap_sort(arr : list[int]):
    # step1: heapify the whole array
    n = len(arr)
    # no. of non leaf nodes = n/2
    # so first non leaf node index from down = (n/2) -1
    for i in range((n//2) - 1, -1, -1):
        heapify(arr, i, n)

    # step 2: swap the first node with the last and call down heapify on the rest part
    for i in range(n - 1, 0, -1):
        # swap
        arr[i], arr[0] = arr[0], arr[i]
        # heapify the first elem so that 0 to( n-1-i) is a heap
        heapify(arr, 0, i)
```

```python
# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
```

### Theory:
### <u>Dijkstras Algorithm:</u><br>
The Dijkstras Algorithm is a greedy algorithm used to find the shortest paths from a source node to all other nodes in a graph with non-negative edge weights.
- Initialisation : <br>
Start by initialising the distance matrix(or priority queue for efficiency) with infinity and the source distance as zero. Also maintain a  boolean array to keep track of the visited and unvisited vertices.

- Algorithm Execution:<br>
1) Begin with the source node and update the distances to its adjacent nodes.<br>
2) Extract the node with the smallest distance from the priority queue.<br>
3) For each neighboring node, calculate the tentative distance from the source node via the current node. If this distance is smaller than the known distance, update the distance and the priority queue.<br>
4) Mark as Visited: After processing the current node and updating distances, mark it as visited.<br>
5) Repeat this process v number of times

- Property<br>
1) Greedy Choice: At each step, the algorithm selects the node with the smallest tentative distance, ensuring that the shortest path to that node is finalized.


#### COMPLEXITY ANALYSIS

- **Time Complexity**: \(O(V^2)\)
  - The algorithm iterates \(V\) times. In each iteration, finding the minimum distance node in a simple array takes \(O(V)\) time. Exploring all neighbors of the minimum distance node also takes \(O(V)\) time in the worst case. Therefore, the total time complexity is \(O(V \times (V + V)) = O(V^2)\).

- **Space Complexity**: \(O(V^2)\)
  - The space required to store the adjacency matrix is \(O(V^2)\). Additionally, the visited array and distance array each take \(O(V)\) space, resulting in a total space complexity of \(O(V^2)\).

#### Using Adjacency List:

- **Time Complexity**: \(O((V + E) \log V)\)
  - Using a priority queue (min-heap) to manage nodes, each node is processed once, and each edge is considered once. The priority queue operations (insert and extract-min) each take \(O(log V)\) time. Thus, the total time complexity is \(O((V + E) log V)\).

- **Space Complexity**: \(O(V + E)\)
  - The space required to store the adjacency list is \(O(V + E)\). The visited array and distance array each take \(O(V)\) space, resulting in a total space complexity of \(O(V + E)\).


#### limitations or conditions for the applicability of Dijkstras algorithm.
- Non-Negative Weights:
Works only with non-negative edge weights. Not suitable for graphs with negative edge weights.

- No negative cycles:
Dijkstras algorithm is sensitive to negative cycles, which are cycles with a total weight less than zero. If a graph contains negative cycles, Dijkstras algorithm will not terminate or will produce incorrect results.

### <u>Bellman-Ford Algorithm</u>

The Bellman-Ford Algorithm is a single-source shortest path algorithm that can handle graphs with negative edge weights and detect negative weight cycles.

**Algorithm:**

- **Initialization**: Set the distance to the source node as zero and all other nodes as infinity.
- **Relaxation**: Iteratively update the distance to each node by checking all edges in the graph.

**Relaxation Process:**

1. **Iteration**: For each edge in the graph, check if the distance to the destination node can be reduced by going through the source node of the edge.
2. **Update**: If the distance can be reduced, update the shortest distance.

   This process is repeated for \(V - 1\) iterations (where \(V\) is the number of vertices) to ensure the shortest paths are found.

**Handling Negative Weight Edges:**

- **Flexibility**: Bellman-Ford can handle graphs with negative edge weights due to its iterative relaxation process, which allows it to correct the effects of negative weights.

### Time and Space Complexity

**1. Using Adjacency Matrix:**

- **Time Complexity**: \(O(V^2)\)
  - Each of the \(V - 1\) iterations involves checking all \(V\) edges, leading to a complexity of \(O(V^2)\).

- **Space Complexity**: \(O(V^2)\)
  - The adjacency matrix requires \(O(V^2)\) space. Additional space for the distance array and other variables is \(O(V)\), so the total space complexity is dominated by the adjacency matrix.

**2. Using Adjacency List:**

- **Time Complexity**: \(O(V x E)\)
  - Each of the \(V - 1\) iterations involves relaxing all \(E\) edges, leading to a complexity of \(O(V x E)\). This is more efficient for sparse graphs compared to the adjacency matrix.

- **Space Complexity**: \(O(V + E)\)
  - The adjacency list requires \(O(V + E)\) space. Additional space for the distance array and other variables is \(O(V)\), making the total space complexity \(O(V + E)\).

### Detection of Negative Weight Cycles

**Detection Process:**

1. **Final Relaxation**: After \(V - 1\) iterations, perform one additional iteration over all edges.
2. **Cycle Check**: If any distance can still be reduced during this extra iteration, a negative weight cycle exists in the graph.

### <u>Floyd-Warshall Algorithm</u>

The Floyd-Warshall Algorithm computes the shortest paths between all pairs of nodes in a graph. It is ideal for dense graphs or scenarios where you need to find shortest paths between every pair of nodes.

**Algorithm:**

1. **Initialization**:
   - Initialize a distance matrix `dist` where each entry `dist[i][j]` represents the shortest distance from node `i` to node `j`. Initially, this distance is set to the weight of the edge between `i` and `j`, or infinity if there is no direct edge.

2. **Algorithm Execution**:
   - Use three nested loops to update the distance matrix. For each pair of nodes `(i, j)`, check if the distance through an intermediate node `k` is shorter than the current known distance:
     \[
     \text{dist}[i][j] = \min(\text{dist}[i][j], \text{dist}[i][k] + \text{dist}[k][j])
     \]
   - Repeat this for all nodes `k`, then for all pairs `(i, j)` to ensure all shortest paths are computed.

**Time Complexity**:

- **Time Complexity**: \(O(V^3)\)
  - The algorithm involves three nested loops, each running \(V\) times (where \(V\) is the number of vertices). Thus, the time complexity is \(O(V^3)\).

**Space Complexity**:

- **Space Complexity**: \(O(V^2)\)
  - The space required to store the distance matrix is \(O(V^2)\). Additional space for the distance matrix and other variables is also \(O(V^2)\).

**Advantages Over Dijkstras and Bellman-Ford**:

- **All-Pairs Shortest Path**:
  - Floyd-Warshall computes shortest paths between all pairs of nodes in a single run, whereas Dijkstras and Bellman-Ford compute shortest paths from a single source node to all other nodes. For all-pairs shortest path problems, Floyd-Warshall is more efficient as it avoids multiple runs of single-source algorithms.

- **Handling Negative Weights**:
  - Like Bellman-Ford, Floyd-Warshall can handle negative edge weights. However, it also detects negative weight cycles, which can be useful for understanding the presence of cycles that reduce path distances indefinitely.

- **Simpler Implementation for Dense Graphs**:
  - Floyd-Warshall is simpler to implement for dense graphs where all-pairs shortest paths are needed, compared to running Dijkstras algorithm \(V\) times or using Bellman-Ford repeatedly.

## Dijkstras Algorithm Implementation

```python
import sys

def get_min_vertex(distance, visited):
    min_vertex = -1
    for i in range(len(distance)):
        if not visited[i] and (min_vertex == -1 or distance[i] < distance[min_vertex]):
            min_vertex = i
    return min_vertex


def dijkstra(adjmatrix, src):
    # initialisation
    v = len(adjmatrix)
    visited = [False] * v
    distance = [sys.maxsize] * v
    distance[src] = 0

    for _ in range(v):
        min_vertex = get_min_vertex(distance, visited)
        visited[min_vertex] = True
        # explore the "unvisited" neighbours
        for i in range(v):
            if adjmatrix[min_vertex][i] > 0 and not visited[i]:
                new_dist = distance[min_vertex] + adjmatrix[min_vertex][i]
                if new_dist < distance[i]:
                    distance[i] = new_dist

    shortest_path = { i : distance[i] for i in range(v)}
    return shortest_path
```

## Bellman-Ford Implementation:

```python
def bellman_ford(graph, start = 0):
    nodes = list(graph.keys())
    distances = {node : float('inf') for node in nodes}
    distances[start] = 0

    # relax the edges v - 1 times
    for _ in range(len(nodes) - 1):
        for u in graph.keys() :
            for v, w in graph[u].items():
                if distances[u] != float('inf') and distances[u] + w < distances[v]:
                    distances[v] = distances[u] + w

    negative_cycle = False
    for u in graph.keys():
        for v, w in graph[u].items():
             if distances[u] != float('inf') and distances[u] + w < distances[v]:
                negative_cycle = True
                break

    return distances, negative_cycle

def print_shortest_paths(distances, is_negative_cycle, start = 0):
    for node, dist in distances.items():
        if not is_negative_cycle:
            print(f"Distance from start {start} to node {node} is {dist}")
        else:
            print("the graph contains negative cycle")
            break
```

## Theory

## Divide and conquer
The Divide and Conquer approach is a problem-solving paradigm that breaks a problem into smaller sub-problems, solves each sub-problem independently, and then combines their solutions to solve the original problem. It is commonly used in algorithms to solve complex problems efficiently.

The three main steps of the Divide and Conquer approach are:

**Divide**: Break the problem into smaller sub-problems that are instances of the same problem.<br>
**Conquer**: Solve each sub-problem independently. If the sub-problems are still large, recursively apply the divide-and-conquer approach to them.<br>
**Combine**: Merge or combine the solutions of the sub-problems to get the solution to the original problem.<br>

## What is the Quick and Heap sort? Formulate it mathematically.
### Quicksort

Quicksort is a highly efficient sorting algorithm that follows the divide and conquer principle. It selects a pivot element from the array, partitions the array into two subarrays (one containing elements less than the pivot and the other containing elements greater than or equal to the pivot), and recursively applies the same procedure to the subarrays.

#### Mathematical Formulation:
Let \( A \) be the array of size \( n \).

- **Divide**: Choose a pivot element \( p \) from the array and partition the array into two parts such that left part contains elements less than pivot and right contains elems greater than pivot

- **Conquer**: Recursively sort left part of pivot array and right part

- **Combine**: Concatenate the sorted parts

#### Time Complexity:
- **Best/Average Case**:  O(nlog n)
- **Worst Case**:  O(n^2)  (when the pivot selection is poor)

---

### Heapsort

Heapsort is a comparison-based sorting algorithm that uses a binary heap data structure (typically a max-heap or min-heap). It transforms the array into a max-heap, then repeatedly extracts the maximum element from the heap and places it in its correct position in the sorted array.

#### Mathematical Formulation:
Let \( A \) be the array of size \( n \).

- **Divide**: Build a max-heap from the array elements. A max-heap is a complete binary tree where each parent node is greater than or equal to its child nodes.

- **Conquer**: Remove the largest element (root of the heap) and place it at the end of the array. Then, heapify the remaining elements to maintain the heap property. Repeat this process until the heap is empty.

- **Combine**: The result is a sorted array in increasing order.

#### Time Complexity:
- **Build heap**: \( O(n) \)
- **Heapify and remove the maximum**: \( O(log n) \) for each element, resulting in \( O(n log n) \) overall.

#### Space Complexity:
- **Heapsort**: \( O(1) \)
- **Quicksort**: \( O(log n)) due to recursion depth.


## Quick Sort

```python
def quickSort(input, s, e):
    if s >= e :
        return
    pivot = getpivot(input, s, e)
    quickSort(input, s, pivot - 1)
    quickSort(input, pivot + 1, e)

def getpivot(input, s, e):
    pivot = e
    j = 0
    for i in range(0, e):
        if(input[i] < input[pivot]):
            # swap
            input[i] , input[j] = input[j], input[i]
            j += 1

    input[e], input[j] = input[j], input[e]
    return pivot
```

```python
arr = [-1, -2, -10 ,100, 20, 70, 1, 2, 20, 3, 10, 29, 4, 5, 0]
quickSort(arr, 0, len(arr) - 1)
print(arr)
```

## Heap Sort

```python
# helper function which downheapifies
# this will heapify the i th position in the array
def heapify(arr: list[int], i: int, n: int):
    parent_index = i
    Lchild_index = 2 * i + 1
    Rchild_index = 2 * i + 2

    while Lchild_index < n:
        min_index = parent_index
        if (arr[Lchild_index] > arr[min_index]):
            min_index = Lchild_index
        if (Rchild_index < n and arr[Rchild_index] > arr[min_index]):
            min_index = Rchild_index
            # return if parent index is the min index
        if (min_index == parent_index):
            return
            # else swap the parent index elem with the min index elem
        arr[parent_index] , arr[min_index] = arr[min_index], arr[parent_index]

        # update the parent_index, Lchild_index and Rchild_index
        parent_index = min_index
        Lchild_index = 2 * parent_index + 1
        Rchild_index = 2 * parent_index + 2


# main function
def heap_sort(arr : list[int]):
    # step1: heapify the whole array
    n = len(arr)
    # no. of non leaf nodes = n/2
    # so first non leaf node index from down = (n/2) -1
    for i in range((n//2) - 1, -1, -1):
        heapify(arr, i, n)

    # step 2: swap the first node with the last and call down heapify on the rest part
    for i in range(n - 1, 0, -1):
        # swap
        arr[i], arr[0] = arr[0], arr[i]
        # heapify the first elem so that 0 to( n-1-i) is a heap
        heapify(arr, 0, i)
```

```python
arr = [100, 20, 70, 1, 2, 20, 3, 10, -1, -11,  29, 4, 5, 0]
heap_sort(arr)
print(arr)
```

## Maximum sum subarray
(Using `divide and conquer`)

### Intuition :<br>
1) Find the subarray from the left part which has the maximum sum<br>
2) Find the subarray from the right part which has the maximum sum<br>
3) Now since the max sum subarray can be a subarray that includes mid , some part of left and some part of right, so find that max sum subarray after these two recursive calls that contains mid, some part of left and some part of right<br>
4) Finally find the max of these three cases and return<br>


### Time and Space Complexity analysis:
Divide : We divide the subarrays into two parts.Thus the maximum depth of recursion call stack in logn. Thus 2 * logn calls will be made at max.<br>

In every level of recursion  we perform a linear scan to compute the maximum crossing sum.<br>
Since the linear scan takes O(n) and happens at each level of recursion, we end up with the complexity of the algorithm being **O(N logN)**

```python
def max_crossing_sum(arr, left, mid, right):
    left_sum = float('-inf')
    sum_left = 0
    start_left = mid

    for i in range(mid, left - 1, -1):
        sum_left += arr[i]
        if sum_left > left_sum:
            left_sum = sum_left
        # Update the starting index of the left subarray
            start_left = i

    right_sum = float('-inf')
    sum_right = 0
    start_right = mid + 1

    for i in range(mid + 1, right + 1):
        sum_right += arr[i]
        if sum_right > right_sum:
            right_sum = sum_right
        # Update the starting index of the right subarray
            start_right = i

    return arr[start_left : start_right + 1]


def max_sum_subarray(arr, left, right):
    # base case
    if left == right:
        return arr[left  : right + 1]

    mid = (left + right) // 2

    # Find maximum subarray sum in left half, right half, and crossing
    leftsubarray = max_sum_subarray(arr, left, mid)
    rightsubarray = max_sum_subarray(arr, mid + 1, right)
    crosssubarray = max_crossing_sum(arr, left, mid, right)

    # Determine the maximum of the three possibilities
    if sum(leftsubarray) >= sum(rightsubarray) and sum(leftsubarray) >= sum(crosssubarray):
        return leftsubarray
    elif sum(rightsubarray) >= sum(leftsubarray) and sum(rightsubarray) >= sum(crosssubarray):
        return rightsubarray
    else:
        return crosssubarray
```

Example usage

```python
arr = [-2, -1, -3, -4, -1, -2, -1, -5, -4]
result = max_sum_subarray(arr, 0, len(arr) - 1)
print("Maximum Subarray Sum:", result)
```
